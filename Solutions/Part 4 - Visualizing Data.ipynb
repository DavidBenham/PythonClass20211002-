```python
# Load libraries
# Load the following libraries: numpy, pandas, matplotlib
import numpy as np
import pandas as pd
#from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
```

```python
# Read .csv into a dataframe called wells
wells = pd.read_csv("../data/enverus_data.csv", low_memory=False)
```

```python
# Build a basic plot of column First12MonthOil_BBL from wells with markers that are "+" signs
# Label the y and x axes
%matplotlib inline
plt.plot(wells["First12MonthOil_BBL"], 'b+')
plt.ylabel("Value")
plt.xlabel("Row Index")
plt.show()
```
